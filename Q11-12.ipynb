```python
import random
import matplotlib.pyplot as plt

times = 1000
dice = [2, 3, 4, 5]

for n in dice:
    averages_list = []
    for i in range(times):
        rolls = [random.randint(1, 6) for j in range(n)]
        average = sum(rolls) / n
        averages_list.append(average)
        
    plt.hist(averages_list, bins=20)
    plt.title(f" {n} Dice")
    plt.xlabel("Average")
    plt.ylabel("Frequency")
    plt.show()
```

```python
import random

def kruskal_count(n_cards, face_steps):
    deck = []
    for i in range(n_cards // 4):
        for j in range(4):
            deck.append(i+1)
    random.shuffle(deck)

    # count the steps
    pos = 0
    for i in range(10):
        card = deck[i]
        if card in [11, 12, 13]:
            pos += face_steps
        else:
            pos += card % 10
    return pos % 2 == 0

n_cards_list = [52, 104]
face_steps_list = [1, 3, 5, 7, 9]

for n_cards in n_cards_list:
    for face_steps in face_steps_list:
        count = 0
        for i in range(10000):
            if kruskal_count(n_cards, face_steps):
                count += 1
        print(f"n_cards={n_cards}, face_steps={face_steps}, probability={count/10000}")
```
